```python
#import libraries
import pandas as pd
import numpy as np
import collections
```

```python
#load dataset
dataset = pd.read_excel("Experiment_1.xlsx")
dataset
```

```python
#pick only the useful columns
df=dataset[['54: TC_Smoke (°C)', '154: TC_Tree_Tr1_060 (°C)', '156: TC_Tree_Tr1_260 (°C)',
                  '157: TC_Tree_Tr2_060 (°C)', '159: TC_Tree_Tr2_260 (°C)', '169: TC_Tree_Tr3_060 (°C)', 
                  '171: TC_Tree_Tr3_260 (°C)', '172: TC_Tree_Tr4_060 (°C)', '174: TC_Tree_Tr4_260 (°C)',
                  '160: TC_Tree_Tr5_060 (°C)', '162: TC_Tree_Tr5_260 (°C)', '175: TC_Tree_Tr6_060 (°C)',
                  '177: TC_Tree_Tr6_260 (°C)', '187: CO_room (%)']]

#convert CO_room in percentage to ppm
df.loc[:,"187: CO_room (ppm)"] = df.loc[:,"187: CO_room (%)"].apply(lambda x : x*10000)

#remove CO_room in percentage from the dataframe
cols = list(df.columns)
cols.remove("187: CO_room (%)")
del df['187: CO_room (%)']

df
```

# Random Selection 

```python
#make a dictionary of robot current position and potential movement
pos_mov= {"Tree_Tr2":["Tree_Tr3","Tree_Tr5","Tree_Tr6"], 
          "Tree_Tr3":["Tree_Tr2","Tree_Tr5","Tree_Tr6"], 
          "Tree_Tr4":["Tree_Tr1","Tree_Tr5","Tree_Tr6"], 
          "Tree_Tr5":["Tree_Tr2","Tree_Tr3","Tree_Tr4","Tree_Tr1","Tree_Tr6"], 
          "Tree_Tr6":["Tree_Tr2","Tree_Tr3","Tree_Tr4","Tree_Tr1","Tree_Tr5"],
          "Tree_Tr1":["Tree_Tr4","Tree_Tr5","Tree_Tr6"]}

#get position
pos = list(pos_mov.keys())
```

```python
#the number of observations selected per movement
desired_rows_data = 5

#check length of dataframe 
data_len = df.shape[0]

#get the number of movement for the robots
no_of_movement = data_len // desired_rows_data

#select only temperature columns. This signifies the different locations in the room
useful_cols2 = list(set(df.columns) - set(['54: TC_Smoke (°C)', '187: CO_room (ppm)']))
```

```python
#create a new dataframe to store the randomly selected readings
new_dataset = collections.defaultdict(list)
pos_list=[]

#randomly select robot's initial position
init_pos = np.random.choice(["Tree_Tr2","Tree_Tr3"])
next_pos = init_pos.copy()
n_rows=0

#loop through all number of movements
for i in range(no_of_movement):     
    
    #collect data 
    for cols in useful_cols2:
        if cols.split("_", 1)[1].startswith(next_pos):
            col_name = cols.split("_")[-1]
            new_dataset[col_name].extend(list((df.loc[:,cols].values)[n_rows:n_rows+5]))
    n_rows+=5
    
    #append robot current position
    pos_list.append(next_pos)
    
    #select robot next position
    next_mov = np.random.choice(pos_mov[next_pos])
    next_pos = next_mov.copy()
```

```python
custom_data = pd.DataFrame(new_dataset)
custom_data[['54: TC_Smoke (°C)', '187: CO_room (ppm)']] = df[['54: TC_Smoke (°C)', '187: CO_room (ppm)']]
custom_data
```

# Data Labeling

```python
#The dataset are labeled into three classes using CO and temperature readings at 60cm
#Because we assume the robot can only capture data at 60cm
#Pick temperature reading at 60cm
cols1 = list(custom_data.columns)
cols1[:1]
```

```python
#include class label into the dataframe
#create an empty dictionary 
df_labels= []

#iterate through the rows of the dataframe
#get the length of the dataframe
df_len = custom_data.shape[0]

for i in range(df_len):
    
    #if any of the values are greater than 600, set label == 2
    filter_rows =[num for num in list(custom_data.iloc[i][cols1[:1]]) if num > 600]   
    if len(filter_rows) > 0 or custom_data.iloc[i]["187: CO_room (ppm)"] > 1600:
        df_labels.append("danger")
        
    elif sum(list(custom_data.iloc[i][cols1[:1]].between(270, 600, inclusive='right'))) or \
        (custom_data.iloc[i]["187: CO_room (ppm)"] > 800 and custom_data.iloc[i]["187: CO_room (ppm)"] <= 1600):
        df_labels.append("caution")
        
    else:
        df_labels.append("safe")
        

#create another column in the original dataframe with the class labels
custom_data.loc[:,"labels"] = df_labels

#save to excel
custom_data.to_excel("Exp1_Labeled_Custom_data.xlsx", index=False)
custom_data
```
